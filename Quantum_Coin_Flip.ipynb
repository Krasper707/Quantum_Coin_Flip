```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(1, 1)

qc.h(0) #.h stands for hadamard gate

qc.measure(0, 0) #measuring it inorder to collapse it

#Circuit simulation
simulator = AerSimulator()
result = simulator.run(qc, shots=1).result()

counts = result.get_counts(qc)
print("Quantum Coin Flip Result:", counts)

print(qc.draw(output='text'))
result_1000 = simulator.run(qc, shots=1000).result()
counts_1000 = result_1000.get_counts(qc)
plot_histogram(counts_1000)
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

def quantum_coin_flip(num_flips):
    fig, axes = plt.subplots(1, num_flips, figsize=(5 * num_flips, 5))

    if num_flips == 1:
        axes = [axes]

    for i in range(num_flips):
        qc = QuantumCircuit(1, 1)

        qc.h(0)

        qc.measure(0, 0)

        simulator = AerSimulator()
        result = simulator.run(qc, shots=1000).result()

        counts = result.get_counts(qc)

        plot_histogram(counts, ax=axes[i])
        axes[i].set_title(f"Coin Flip {i + 1}")

    plt.tight_layout()
    plt.show()

```

```python
quantum_coin_flip(2)
```
